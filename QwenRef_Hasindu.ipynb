```python
from transformers import AutoModelForCausalLM, AutoTokenizer

model_name = "Qwen/Qwen2.5-3B-Instruct"
tokenizer = AutoTokenizer.from_pretrained(model_name, trust_remote_code=True)
model = AutoModelForCausalLM.from_pretrained(model_name,
                                             device_map="auto",
                                             trust_remote_code=True)
```

```python
import torch

# Example prompt
prompt = "Did you trained with research papers?"

# 1. Tokenize the prompt
input_ids = tokenizer.encode(prompt, return_tensors="pt")

# 2. Move input to the same device as the model (if using GPU)
input_ids = input_ids.to(model.device)

# 3. Generate output
with torch.no_grad():
    output_ids = model.generate(
        input_ids=input_ids,
        max_new_tokens=150,   # Limit the length of the generated text
        temperature=0.7,      # Controls "creativity" of the output
        top_p=0.9,            # Controls sampling diversity
        do_sample=True        # Enables sampling (instead of greedy decoding)
    )

# 4. Decode the generated tokens to text
response = tokenizer.decode(output_ids[0], skip_special_tokens=True)
print(response)
```

```python
import os
import fitz  # PyMuPDF
import markdown
import re

# ✅ Corrected filenames to match your `ls /content` output
pdf_path = "/content/2501.12948v1.pdf"
md_files = [
    "/content/dataset.md",
    "/content/deepseekv3-cost-explained.md",  # ✅ Corrected filename
    "/content/deepseekv3-explained.md",       # ✅ Corrected filename
    "/content/design-notes-3fs.md",
    "/content/open-source-week.md"
]

def extract_text_from_pdf(pdf_file):
    text = []
    with fitz.open(pdf_file) as doc:
        for page in doc:
            text.append(page.get_text())
    return "\n".join(text)

def extract_text_from_markdown(md_file):
    with open(md_file, "r", encoding="utf-8") as f:
        md_content = f.read()
    html = markdown.markdown(md_content)  # Convert Markdown to HTML
    text = re.sub(r'<[^>]+>', '', html)  # Strip out HTML tags
    return text

# Parse PDF
pdf_text = extract_text_from_pdf(pdf_path)

# Parse each Markdown file
md_texts = [extract_text_from_markdown(mdf) for mdf in md_files]

# Combine everything into a single corpus (optional)
full_corpus = pdf_text + "\n" + "\n".join(md_texts)

print("PDF length (chars):", len(pdf_text))
print("Combined MD length (chars):", sum(len(t) for t in md_texts))
```

```python
import random

keywords = ["technique", "method", "result", "conclusion", "analysis", "design"]
def naive_qa_generator(text, num_questions=10):
    sentences = text.split(".")
    qa_pairs = []
    for sentence in sentences:
        sentence = sentence.strip()
        for kw in keywords:
            if kw in sentence.lower():
                # Create a naive question
                question = f"What does this document say about {kw}?"
                answer = sentence + "."
                qa_pairs.append({"question": question, "answer": answer})
                break
    # Randomly sample up to num_questions
    return random.sample(qa_pairs, min(num_questions, len(qa_pairs)))

# Example: Generate Q&A from PDF and one MD file
pdf_qa = naive_qa_generator(pdf_text, num_questions=20)
md_qa = naive_qa_generator(md_texts[0], num_questions=20)

# Combine them
all_qa = pdf_qa + md_qa

print("Sample Q&A:", all_qa[:3])
```

```python
from sklearn.model_selection import train_test_split

train_qa, temp_qa = train_test_split(all_qa, test_size=0.2, random_state=42)
val_qa, test_qa = train_test_split(temp_qa, test_size=0.5, random_state=42)

print(f"Train size: {len(train_qa)}, Val size: {len(val_qa)}, Test size: {len(test_qa)}")

# Optionally convert to Hugging Face Dataset
from datasets import Dataset

train_dataset = Dataset.from_list(train_qa)
val_dataset = Dataset.from_list(val_qa)
test_dataset = Dataset.from_list(test_qa)
```
